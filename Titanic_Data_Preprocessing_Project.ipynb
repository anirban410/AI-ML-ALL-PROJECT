# Titanic Data Preprocessing Project

## Project Description
This project demonstrates a comprehensive data preprocessing pipeline for the classic Titanic dataset. The goal is to clean, transform, and prepare the raw passenger data for machine learning model training. The notebook covers essential steps including data loading, handling missing values, encoding categorical features, and scaling numerical features.

## Project Steps & Topics Covered
This project systematically addresses key data preprocessing challenges, covering the following topics:

1.  **Data Fetching:** Loading the Titanic dataset directly from a GitHub raw CSV URL.
2.  **Data Cleaning:** Initial exploration to understand data structure, shape, and column information.
3.  **Handling Missing Values:** Strategies for imputing missing data in 'Age' (using median) and 'Embarked' (using mode), and identifying columns with excessive missing values.
4.  **Feature Engineering (Implicit):** Dropping unnecessary columns like `PassengerId`, `Name`, and `Ticket` to reduce noise and dimensionality.
5.  **One-Hot Encoding:** Converting nominal categorical features ('Sex', 'Embarked') into a numerical format suitable for machine learning algorithms.
6.  **Label Encoding:** Transforming the target variable ('Survived') into a numerical format.
7.  **Train-Test Split:** Dividing the dataset into training and testing sets to evaluate model performance objectively.
8.  **Standardization:** Applying `StandardScaler` to numerical features to standardize their ranges.
9.  **Normalization:** Applying `MinMaxScaler` to numerical features to normalize their ranges (scaling to a 0-1 range).

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
```

```python
#step 2: load the CSV FILE FROM GIT HUB

csv_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(csv_url)
df
```

```python
#   STEP 3: PRINT FIRST 5 ROWS
print ("first 5 row of the data set")
display(df.head())
```

```python
# STEP 4: CHECK THE SHAPE OF THE DATA SET

print(df.shape)
```

```python
#STEP 5: DISPLAY ALL COLUMN NAMES

print ("\nColumn names")
print (df.columns)
```

```python
# STEP 6: CHEECK BASIC INFORMATION ABOUT THE DATASET

print("\nDataset info:")
df.info()
```

```python
#STEP 7: CHECK MISSING VALUES

print("\nMissing values before cleaning:")
print(df.isnull().sum())
```

```python
#STEP 8: DROP UNNECESSARY COLUMNS

drop_cols = ['PassengerId', 'Name', 'Ticket', 'cabin']
existing_drop_cols = [col for col in drop_cols if col in df.columns]
df.drop(columns=existing_drop_cols, inplace=True)
print ("\n columns after dropping unnecessary features:")
print (df.columns)
```

```python
# STEP 9: HANDLE MISSING VALUES

# Fill missing values in Age with the median
# Median is used because it is better than mean when outliers may exist
if "Age" in df.columns:
    df["Age"] = df["Age"].fillna(df["Age"].median())

if "Embarked" in df.columns:
    df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
#STEP 10: SEPARATE FEATURES (X) AND TARGET (Y)

x = df.drop(columns=["Survived"])
y = df["Survived"]

print ("\nFeatures (X) preview:")
display(x.head())

print ("\nTarget (Y) preview:")
display(y.head())
```

```python
# STEP 11: APPLY ONE SHOT ENDING TO CATEGORICAL INPUT COLMN
# STEP 11: APPLY ONE-HOT ENCODING TO CATEGORICAL INPUT COLUMNS

categorical_cols = [col for col in ["Sex", "Embarked"] if col in x.columns]

x = pd.get_dummies(x, columns=categorical_cols, drop_first=True)

print("\nData after one-hot encoding:")
display(x.head())

print("\nColumns after one-hot encoding:")
print(x.columns)
```

```python
# STEP 12: APPLY LABEL ENCODING TO THE TARGET COLUMN
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
print("\nEncoded target values:")
print(y_encoded[:10])
```

```python

# STEP 13: SPLIT DATA INTO TRAINING AND TESTING SETS

# train_test_split divides the data into:
# - training data: used to train the model
# - testing data: used to evaluate the model

X_train, X_test, y_train, y_test = train_test_split(
    x, y_encoded, test_size = 0.2, random_state=42
)

print("\nTrain-Test Split Shapes:")
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
#STEP 14: APPLY STANDARDIZATION

if 'Cabin' in X_train.columns:
    X_train = X_train.drop(columns=['Cabin'])
if 'Cabin' in X_test.columns:
    X_test = X_test.drop(columns=['Cabin'])

standard_scaler = StandardScaler()

x_train_standardized = standard_scaler.fit_transform(X_train)
x_test_standardized = standard_scaler.transform(X_test)

print("\nfirst 5 rows after standardization:")
print(x_train_standardized[:5])
```

```python
# STEP 15: APPLY NORMALIZATION

minmax_scaler = MinMaxScaler()

x_train_normalized = minmax_scaler.fit_transform(X_train)
x_test_normalized = minmax_scaler.transform(X_test)
print ("\nfirst 5 rows after normalization:")
print (x_train_normalized[:5])
```
